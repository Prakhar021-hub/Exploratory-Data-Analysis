```python
case_no_index = df.columns.get_loc("case_no")
case_number_index = df.columns.get_loc('case_number')
print(case_no_index)
print(case_number_index)
```

```python
case_number_list = []
for i in df.iloc[0:135269, case_no_index]:
    case_number_list.append(i)
for i in df.iloc[135269:374362, case_number_index]:
    case_number_list.append(i)
```

```python
df['case_numbers'] = case_number_list
df['case_numbers'].nunique()
```

```python
df.drop(['case_no', 'case_number'], axis=1, inplace=True)
```

```python
df['case_status'].value_counts() 

# from the data we can see that 18194 people have withdrawn there application => those data dosent make any sense 
# also we will use this 'case_status' column as our target column (thus lets clean this column)
# also we can combine the data 'certified' and 'certified-Expired' as they represents certified 
```
